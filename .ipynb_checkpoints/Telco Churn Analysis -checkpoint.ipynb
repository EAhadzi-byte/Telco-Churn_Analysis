# Customer Churn Analysis & Prediction in TELCO Sector

## Project Overview
This project aim at identify customers who will be canceling their subscription or failing to renew within a set period.
   
   - Voluntary Churn: Customer that decides to leave(dissatisfaction, price).
   - Involuntary Churn: Paymenent failure or expired credit cards.
   
Why Churn matters:
  - Customer Lifetime Value(CLV): High churn decreases total revenue overtime.
  - Helps in planning company growth.
  - it acts as a feedback loop. if churn is high the product is not solving the problem.
  - Cost efficience. Retention is cheaper than acquiring new customers
   

```python
#Importing neccessay libraries
import io
import requests

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Useful functions for accessing remote files stored on github

def get_raw_github_url(url: str) -> str:
    """
    Converts a standard GitHub file URL into a raw URL.
    Example: 
    'https://github.com/user/repo/blob/main/data.csv' 
    -> 'https://raw.githubusercontent.com/user/repo/main/data.csv'
    """
    if 'raw.githubusercontent.com' in url:
        return url
    
    return url.replace('github.com', 'raw.githubusercontent.com').replace('/blob/', '/')

def load_github_data(url: str) -> pd.DataFrame:
    """
    Loads a CSV from a public GitHub repo into a DataFrame.
    """
    raw_url = get_raw_github_url(url)
    try:
        df = pd.read_csv(raw_url)
        print(f"Successfully loaded data from: {raw_url}")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None
```

```python
# Reading data set
url = 'https://github.com/EAhadzi-byte/Churn_Analysis/blob/main/WA_Fn-UseC_-Telco-Customer-Churn%202.csv'

df = load_github_data(url)

df.head()
```

# *Data Cleaning and Preparation*

```python
# Finding all the necessary information about the data
df.info()
```

- From the data we can notice that there are 7,043 rows and 21 columns.
- There are 3 numeric columns and 18 object (categorical) columns.
- TotalCharges is an object but should be a float.
- Churn is currently an 'object'('yes'/'No'). Models need this to be 1 or 0.

```python
# Fix the TotalCharges object-to-float 
df['TotalCharges'] = df['TotalCharges'].replace(' ',np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

#Check how many NaNs we created
df['TotalCharges'].isnull().sum()
```

```python
#Fill missing TotalCharges values with 0, low tenure (new_customers)
df[df["TotalCharges"].isnull()][["tenure", 
                                 "MonthlyCharges", 
                                 "TotalCharges"]].head()
```

```python
# Convert Churn to binary
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['Churn'].value_counts()
```

```python
# Checking for duplicate
df.duplicated().sum()
```

```python
df.info()
```

## EXPLORATORY DATA ANALYSIS (EDA)

EDA helps us to uncovers and answers deep questions about our data though visualisation before any model is built.

**Question 1. How do churn customers behave differently?**

```python
# A count plot of churn to understand class balance
plt.figure(figsize=(6,4))
sns.countplot(x='Churn'
              , data=df
              , hue= 'Churn'
              , palette='magma'
              )
plt.title('Distribution of Churn vs Non-Churn Customers')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

ANSWER: Churn forms minority of the data set.
 - The churn customers are mostly short-tenure users.
 - They are users on the month to month plan.
 - Paying above-average monthly fees

**Question 2. Does the length of Tenure has impart on Churn?**

```python
#Boxplot showing the distribution of tenure for each churn category
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='tenure'
            ,data=df
            ,hue='Churn'
            ,palette='Set2')
plt.title('Customer Tenure by Churn Status')
plt.xlabel("Churn")
plt.ylabel("Tenure (Months)")
plt.show()
```

**ANSWER:** Yes this tell us that customers with lower tenue churn.
Customers leave almost in the first 0-6 months.
- This signals product value is not realized quickly enough.
- New users are at high risk to churn.

**Question 3. Are we overcharging out customers into leaving?**

```python
# Boxplotting showing the distribution of monthly charges wih churn.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn'
            , y='MonthlyCharges'
            , data=df)
plt.title('Monthly Charges vs Churn')
plt.show()
```

**ANSWER:** Yes, Higher prices are driving customers away.
- Because customers might not be getting the expected value.
- Premium users have high expectation and low tolerance to services issues.
  

**Question 4. Does Online Security have different churn patterns?**

```python
# A count plot of Online Security vs churn
plt.figure(figsize=(8,5))
sns.countplot(x='OnlineSecurity'
              , hue="Churn"
              , data=df
              , palette='viridis')

plt.title('Impact of Online Security on Customers')
plt.xlabel('Has Online Security')
plt.ylabel('Number of Customers')
plt.legend(title='Churned')
plt.show()
```

**ANSWER:** Yes

Customers who don't have online security mostly churn.
- Lower switching costs. Because customers don't feel safe and easy to move to competitors.
- Customers without Tech Support are more likly to Churn instead of getting help they get frustrated and cancel.
- Customers with high-speed Fiber churn more if sell them Fiber without security or a Tech Support. THey have low tolerance.

**Question 5. Are Certain payment methods linked to higher exit rates?**

```python
# Count plot for Churn Rate by Payment Method
sns.countplot(
    data=df, 
    x='PaymentMethod', 
    hue='Churn', 
    palette='magma')
plt.title('Churn Rate by Payment Method')
plt.xticks(rotation=45)
# Adjusting layout to prevent clipping of labels
plt.tight_layout()
plt.show()
```

**ANSWER:** Yes, Electronic Check payment method has the high Churn rate.
- Customers manually evaluation of the service against the cost every 30days.
- Payment failure causing involuntary churn.
- Unwillingness to always manually pay unlike automatic payment.

**Question 6. Which of the features has the stongest mathematical link to Churn?**

```python
#Correlation Heatmap 
plt.figure(figsize=(6,4))
sns.heatmap(df[["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges", "Churn"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

What we notice:
- Contract_Month-to-month: This usually has the highest positive correlation.
- Tenure: This has a negative correlation.(The longer the stay the less chance they will leave)
- Fiber users churn more. Is the fiber service unreliable or too expensive?
